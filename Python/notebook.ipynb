```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./../Datasets/df_anscombe.csv')
df.head()
```

```python
# Crear una figura con 4 subplots (2 filas x 2 columnas)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
grupos = sorted(df['group'].unique())

for i, grupo in enumerate(grupos):
    ax = axs[i // 2, i % 2]
    data = df[df['group'] == grupo]
    
    # Graficar puntos
    ax.scatter(data['x'], data['y'], color='red', s=50)
    
    # # Calcular la lnea de regresin (ajuste lineal)
    # slope, intercept = np.polyfit(data['x'], data['y'], 1)
    # x_vals = np.array([data['x'].min(), data['x'].max()])
    # y_vals = intercept + slope * x_vals
    # ax.plot(x_vals, y_vals, color='red', lw=2)
    
    # Personalizacin del grfico
    ax.set_title(f'Grupo {grupo}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Definir colores para cada grupo
colores = {1: 'blue', 2: 'orange', 3: 'green', 4: 'red'}

plt.figure(figsize=(10, 8))

# Graficar los puntos de cada grupo con transparencia para visualizar mejor la superposicin
for grupo, color in colores.items():
    data = df[df['group'] == grupo]
    plt.scatter(data['x'], data['y'], color=color, label=f'Grupo {grupo}', s=50, alpha=0.7)

# Superponer un hexbin para mostrar la densidad de puntos de todos los grupos
plt.hexbin(df['x'], df['y'], gridsize=25, cmap='Greys', alpha=0.3, edgecolors='none')

# Calcular la lnea de regresin global (para todos los datos)
slope, intercept = np.polyfit(df['x'], df['y'], 1)
x_vals = np.array([df['x'].min(), df['x'].max()])
y_vals = intercept + slope * x_vals
plt.plot(x_vals, y_vals, color='purple', lw=2, label='Lnea de regresin')

# Personalizacin del grfico
plt.xlabel('x')
plt.ylabel('y')
plt.title("Anscombe's Quartet: Puntos, densidad y lnea de regresin global")
plt.grid(True)
plt.legend()
plt.show()
```
